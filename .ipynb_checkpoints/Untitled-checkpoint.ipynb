```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_phase_space_trajectory(df, cpc_code):
    """
    Generates a Phase Space Scatter Plot (Velocity vs. Acceleration) 
    for a specific CPC Technology code.
    """
    # ---------------------------------------------------------
    # 1. PREPROCESS DATA
    # ---------------------------------------------------------
    # Filter for the specific CPC
    df_cpc = df[df['cpc_group'] == cpc_code].copy()
    
    # Set index to Year and ensure numeric sorting
    df_cpc = df_cpc.set_index('innovation_year').sort_index()
    
    # Fill missing years with 0 to ensure continuity of the curve
    all_years = range(df_cpc.index.min(), df_cpc.index.max() + 1)
    df_cpc = df_cpc.reindex(all_years, fill_value=0)

    # ---------------------------------------------------------
    # 2. CALCULATE VECTORS (KINEMATICS)
    # ---------------------------------------------------------
    # SMOOTHING: 3-Year Rolling Mean (Standard for bibliometrics)
    # This removes spikes from single massive portfolio filings
    counts_smooth = df_cpc['family_count'].rolling(window=3, center=True).mean()
    
    # VELOCITY (v): The rate of new filings (First Derivative)
    velocity = counts_smooth.diff()
    
    # ACCELERATION (a): The change in rate (Second Derivative)
    acceleration = velocity.diff()

    # Create a clean plotting dataframe (removing NaNs from the diff/rolling steps)
    plot_data = pd.DataFrame({
        'Year': velocity.index,
        'Velocity': velocity,
        'Acceleration': acceleration
    }).dropna()

    # ---------------------------------------------------------
    # 3. PLOT THE TRAJECTORY
    # ---------------------------------------------------------
    plt.figure(figsize=(10, 8))
    sns.set_style("whitegrid")
    
    # SCATTER: Plot one point per year, colored by time
    points = plt.scatter(
        plot_data['Velocity'], 
        plot_data['Acceleration'], 
        c=plot_data['Year'], 
        cmap='viridis',    # Color Map: Yellow (Old) -> Purple (New)
        s=120,             # Size of dots
        edgecolor='k', 
        alpha=0.8,
        zorder=3
    )

    # TRACE: Connect the dots to show the path
    plt.plot(
        plot_data['Velocity'], 
        plot_data['Acceleration'], 
        color='gray', 
        alpha=0.4, 
        linewidth=2, 
        zorder=2
    )

    # ANNOTATIONS: Mark Start and End Years
    if not plot_data.empty:
        # Start Year
        plt.text(plot_data['Velocity'].iloc[0], plot_data['Acceleration'].iloc[0]+0.5, 
                 str(int(plot_data['Year'].iloc[0])), fontsize=10, fontweight='bold')
        # End Year
        plt.text(plot_data['Velocity'].iloc[-1], plot_data['Acceleration'].iloc[-1]+0.5, 
                 str(int(plot_data['Year'].iloc[-1])), fontsize=10, fontweight='bold', color='purple')

    # QUADRANTS: Add Zero-Lines
    plt.axhline(0, color='black', linestyle='--', linewidth=1, alpha=0.5)
    plt.axvline(0, color='black', linestyle='--', linewidth=1, alpha=0.5)

    # LABELS
    plt.title(f"Innovation Phase Space: {cpc_code}", fontsize=16)
    plt.xlabel("Velocity (New Families / Year)", fontsize=12)
    plt.ylabel("Acceleration (Change in Growth Rate)", fontsize=12)
    plt.colorbar(points, label='Year')
    
```

```python
df = pd.read_csv('innovation_velocity_2003-2024_calibration.csv')
plot_phase_space_trajectory(df, 'G06N')
plt.tight_layout()
plt.show()
```
